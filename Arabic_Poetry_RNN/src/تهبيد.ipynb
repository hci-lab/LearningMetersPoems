```python
# District
cites = {'Budapest': [ 'Terézváros', 
                      'Újbuda', 
                      'Reformed'],
         
         'Debrecen': [ 'Debrecen',
                      'Hajdúsámson', 
                      'omdiano' ],
         
         'Szeged': [ 'Alsózsolca',
                     'Berzék',
                     'Bőcs'],
         
         'Miskolc': ['Egerág', 
                     'Görcsöny', 
                     'Husztót', 
                     'Áta'],
         
         'Pecs': [ 'Józsefváros', 
                   'Ferencváros', 
                   'Quarry']}
i = 0
ids_district = []
districts = []
city = []
for x  in cites:
    print(cites[x])
    for y in cites[x]:
        i += 1
        ids_district.append(i)
        districts.append(y)
        city.append(x)

        
district_dict = {'id': ids_district,
                 'district': districts,
                 'city': city
                }


district_df = pd.DataFrame(district_dict)

district_df
```

```python
from dateutil import rrule, parser

date1 = '2000-05-03'
date2 = '2002-9-20'

dates = list(rrule.rrule(rrule.DAILY,
                         dtstart=parser.parse(date1),
                         until=parser.parse(date2)))

dates_ = []
for i in range(1, 50000):
    dates_.append(str(dates[np.random.randint(1, len(dates))]).split(" ")[0])
```
